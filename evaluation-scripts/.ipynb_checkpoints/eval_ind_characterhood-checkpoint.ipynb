```python
import re
import os
import pandas as pd
import seaborn as sns
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
results_dirs = ['../thesis-university-impl/results/simulation_2023-05-17 15:58:22',
                '../thesis-university-impl/results/simulation_2023-05-17 16:50:54',
                '../thesis-university-impl/results/simulation_2023-05-17 17:02:55',
                '../thesis-university-impl/results/simulation_2023-05-17 17:26:54',
                '../thesis-university-impl/results/simulation_2023-05-17 17:30:26']
```

```python
needs_dir = '/needs'
metadata_path = '/meta.csv'
traces_path = '/traces.txt'
```

```python
actions = ["Inactive", "Walking", "PlayingGame", "Eating", "DrinkingCoffee", 
           "Studying", "Working", "WaitingLecture", "AttendingLecture", "GivingLecture"]
```

```python
actionLogKeys = ["name", "components", "duration", "t1", "tn"]
```

```python
def sample_time_to_real(sample_time, time_step):
    return (time_step*sample_time) / (60*60)
```

```python
def read_csv_files(directory):
    dfs = []

    for filename in os.listdir(directory):
        if filename.endswith('.csv'):
            file_path = os.path.join(directory, filename)
            df = pd.read_csv(file_path)
            dfs.append(df)

    return dfs
```

```python
def parse_behavioral_traces(file_path):
    traces = []
    
    with open(file_path, 'r') as file:
        for line in file:
            line = line.strip()
            trace = []
            matches = re.findall(r'\[([\w\d]+)\(([\w\d\s.,]*)\)](\d+)', line)

            start_time = 0
            for match in matches:
                action_name = match[0]
                components = match[1].split(',') if match[1] else []
                duration = int(match[2])
                
                end_time = start_time + duration - 1
                action = {
                    'name': action_name,
                    'components': components,
                    'duration': duration,
                    'start': start_time,
                    'end': end_time
                }
                
                trace.append(action)
                start_time = end_time + 1
            
            traces.append(trace)
    
    return traces
```

```python
# Read data 
metadata = pd.DataFrame([])
traces = []

for results in results_dirs:
    meta = pd.read_csv(results + metadata_path)
    tcs = parse_behavioral_traces(results + traces_path)
    
    metadata = pd.concat([metadata, meta], ignore_index=True)
    traces.extend(tcs)
```

```python
# New dataframe to contain duration of each action
new_metadata = pd.DataFrame.copy(metadata)
for action in actions:
    new_metadata[action] = 0
new_metadata['total_time_active'] = 0
new_metadata.at['sequence_length'] = 0
```

```python
# Calculate duration for each action 
for i in range(0, len(traces)):
    for action in traces[i]:
        action_name = action['name']
        duration = action['duration']
        
        new_metadata.at[i, action_name] += duration
        new_metadata.at[i, 'total_time_active'] += duration if action_name != "Inactive" else 0
    new_metadata.at[i, 'sequence_length'] = len(traces[i])
```

```python
students_metadata = new_metadata[new_metadata['type'] == 'student'].copy().drop(['workaholic', "Working", "GivingLecture"], axis=1)
admin_metadata = new_metadata[new_metadata['type'] == 'admin'].copy().drop(['driven', "Studying", "AttendingLecture"], axis=1)
professors_metadata = new_metadata[new_metadata['type'] == 'professor'].copy().drop(['driven', "Studying", "AttendingLecture"], axis=1)
```

```python
def assign_driven_group(row):
    value = row['driven']
    if -1 <= value <= -0.75:
        return 'very_low'
    elif -0.75 < value <= -0.25:
        return 'low'
    elif -0.25 < value <= 0.25:
        return 'moderate'
    elif 0.25 < value <= 0.75:
        return 'high'
    elif 0.75 < value <= 1:
        return 'very_high'
    else:
        return 'unknown'
```

```python
def assign_workaholic_group(row):
    value = row['workaholic']
    if -1 <= value <= -0.75:
        return 'very_low'
    elif -0.75 < value <= -0.25:
        return 'low'
    elif -0.25 < value <= 0.25:
        return 'moderate'
    elif 0.25 < value <= 0.75:
        return 'high'
    elif 0.75 < value <= 1:
        return 'very_high'
    else:
        return 'unknown'
```

```python
def remove_outliers_by_group(data, variable, group_variable):
    cleaned_data = pd.DataFrame(columns=data.columns)
    
    # Iterate over unique groups
    for group in data[group_variable].unique():
        group_data = data[data[group_variable] == group]
        q1 = group_data[variable].quantile(0.25)
        q3 = group_data[variable].quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr
        cleaned_group_data = group_data[(group_data[variable] >= lower_bound) & (group_data[variable] <= upper_bound)]
        cleaned_data = pd.concat([cleaned_data, cleaned_group_data])
    
    return cleaned_data
```

```python
admin_metadata['workaholic_group'] = admin_metadata.apply(assign_workaholic_group, axis=1)
admin_metadata['time_productive'] = admin_metadata['Working'] / admin_metadata['total_time_active']

#admin_metadata = remove_outliers_by_group(admin_metadata, 'time_productive', 'workaholic_group')
```

```python
students_metadata['driven_group'] = students_metadata.apply(assign_driven_group, axis=1)
students_metadata['time_productive'] = students_metadata['Studying'] / students_metadata['total_time_active']

#students_metadata = remove_outliers_by_group(students_metadata, 'time_productive', 'driven_group')
```

```python
professors_metadata['workaholic_group'] = professors_metadata.apply(assign_workaholic_group, axis=1)
professors_metadata['time_productive'] = professors_metadata['Working'] / professors_metadata['total_time_active']

#professors_metadata = remove_outliers_by_group(professors_metadata, 'time_productive', 'workaholic_group')
```

```python
correlation, p_value = stats.spearmanr(students_metadata['driven'], students_metadata['time_productive'])
print('Students')
print("Correlation coefficient:", correlation)
print("P-value:", p_value)

correlation, p_value = stats.spearmanr(professors_metadata['workaholic'], professors_metadata['time_productive'])
print('\nProfessors')
print("Correlation coefficient:", correlation)
print("P-value:", p_value)

correlation, p_value = stats.spearmanr(admin_metadata['workaholic'], admin_metadata['time_productive'])
print('\nAdmin')
print("Correlation coefficient:", correlation)
print("P-value:", p_value)
```

p-value is less than 0.05 so there is a correlation
