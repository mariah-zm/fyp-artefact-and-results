```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def plot_needs(dataframe):    
    # Iterate over each column in the data frame
    for column in dataframe.columns:
        sns.lineplot(data=dataframe[column], label=column)

    
    plt.ylabel('Needs Level')
    plt.xlabel('Time Step (t)')

    plt.ylim(0, 1)

    sns.set_style("darkgrid")
    sns.set_palette("colorblind")
    
    plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left')
    plt.show()
```

```python
def read_csv_files(directory):
    dfs = []

    for filename in os.listdir(directory):
        if filename.endswith('.csv'):
            file_path = os.path.join(directory, filename)
            df = pd.read_csv(file_path)
            dfs.append(df)

    return dfs
```

```python
results_dir = '../thesis-university-impl/results/simulation_2023-05-24 19:28:39'
```

```python
needs_dir = results_dir + '/needs'
metadata_path = results_dir + '/meta.csv'
positions_path = results_dir + '/positions.txt'
traces_path = results_dir + '/traces.txt'
```

```python
# Read data 
needs = read_csv_files(needs_dir)
metadata = pd.read_csv(metadata_path)
```
