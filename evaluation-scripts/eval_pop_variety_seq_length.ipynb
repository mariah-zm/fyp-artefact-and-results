```python
def sample_time_to_real(sample_time, time_step):
    return (time_step*sample_time) / (60*60)
```

```python
def read_csv_files(directory):
    dfs = []

    for filename in os.listdir(directory):
        if filename.endswith('.csv'):
            file_path = os.path.join(directory, filename)
            df = pd.read_csv(file_path)
            dfs.append(df)

    return dfs
```

```python
def count_action_length(file_path):
    lengths = []
    
    with open(file_path, 'r') as file:
        for line in file:
            cnt = line.count('[')
            lengths.append(cnt)
    
    return lengths
```

```python
# Read data 
metadata = pd.DataFrame([])
trace_lengths = []

for results in results_dirs:
    meta = pd.read_csv(results + metadata_path)
    tcs = count_action_length(results + traces_path)
    
    metadata = pd.concat([metadata, meta], ignore_index=True)
    trace_lengths.extend(tcs)
```

```python
# New dataframe to contain duration of each action
new_metadata = pd.DataFrame.copy(metadata)
new_metadata['sequence_length'] = trace_lengths
new_metadata['sequence_length'] = new_metadata['sequence_length'] - 2
```

```python
# Visual Detection for outliers
plt.figure()
p = new_metadata.boxplot(column='sequence_length', return_type='axes')
p.set_title('Outliers in Action Sequence Length', pad=10)
p.set_xlabel('Action Sequence Length', labelpad=10)
plt.savefig('seq_length_outliers.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
def determine_outlier_thresholds_iqr(dataframe, col_name, th1=0.25, th3=0.75):
    quartile1 = dataframe[col_name].quantile(th1)
    quartile3 = dataframe[col_name].quantile(th3)
    iqr = quartile3 - quartile1
    upper_limit = quartile3 + 1.5 * iqr
    lower_limit = quartile1 - 1.5 * iqr
    return lower_limit, upper_limit
```

```python
lower, upper = determine_outlier_thresholds_iqr(new_metadata, 'sequence_length')
lower, upper
```

```python
metadata_no_outliers = new_metadata[new_metadata['sequence_length'] > 1]
metadata_no_outliers = metadata_no_outliers[metadata_no_outliers['sequence_length'] < upper]
len(metadata_no_outliers)
```

```python
# Calculating action sequence length
sns.set_style("darkgrid")
sns.histplot(data=metadata_no_outliers, x='sequence_length', kde=True, color='purple')

plt.title('Histogram for Action Sequence Length')

plt.ylabel('Count', labelpad=10)
plt.xlabel('Action Sequence Length',  labelpad=10)
plt.title('Frequency Distribution for Action Sequence Length',  pad=10)
plt.savefig('action_seq_length.png', dpi=300, bbox_inches='tight')
plt.show()
```
